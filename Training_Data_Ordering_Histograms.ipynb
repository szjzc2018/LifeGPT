```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the data from the uploaded file
data_1 = pd.read_csv('LifeGPT//conway_states_0.5_0.5_10000by32by32by10_toroidal_20240813_224815_sorder0.5_eorder0.5.csv')
data_2 = pd.read_csv('LifeGPT//conway_states_0_1_10000by32by32by10_toroidal_20240711_133408.csv')

# Define the order parameter function
def order_parameter(string):
    ones = string.count('1')
    total = len(string)
    return ones / total

# Calculate the order parameter for each row in "State 1" for both datasets
data_1['Order Parameter'] = data_1['State 1'].apply(order_parameter)
data_2['Order Parameter'] = data_2['State 1'].apply(order_parameter)

# Apply the style similar to the recent script
sns.set(style="whitegrid", rc={
    'figure.figsize': (7, 3.35),  # Adjusted for two subplots side by side
    'figure.dpi': 300,
    'lines.linewidth': 2.0,
    'axes.spines.left': True,
    'axes.spines.bottom': True,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'axes.edgecolor': 'black',
    'axes.linewidth': 2.0,
    'xtick.direction': 'out',
    'ytick.direction': 'out',
    'xtick.major.width': 2,
    'ytick.major.width': 2,
    'xtick.major.size': 6,
    'ytick.major.size': 6,
    'axes.labelsize': 10,
    'axes.titlesize': 10,
    'legend.fontsize': 10,
    'legend.title_fontsize': 10,
    'legend.loc': 'best',
    'axes.grid': False
})

# Define the Plasma palette
plasma_palette = sns.color_palette("plasma", 1)

# Create the figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 3.35))

# Plot the first distribution
sns.histplot(data_1['Order Parameter'], bins=30, kde=True, color=plasma_palette[0], linewidth=2, alpha=0.5, ax=ax1)
ax1.set_xlabel('$/eta$', fontsize=10)
ax1.set_ylabel('Frequency', fontsize=10)
ax1.text(-0.1, 1.1, 'A', transform=ax1.transAxes, fontsize=12, fontweight='bold', va='top', ha='right')

# Plot the second distribution
sns.histplot(data_2['Order Parameter'], bins=30, kde=True, color=plasma_palette[0], linewidth=2, alpha=0.5, ax=ax2)
ax2.set_xlabel('$/eta$', fontsize=10)
ax2.set_ylabel('Frequency', fontsize=10)
ax2.text(-0.1, 1.1, 'B', transform=ax2.transAxes, fontsize=12, fontweight='bold', va='top', ha='right')

# Customize the spines for both plots
for ax in [ax1, ax2]:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for spine in ax.spines.values():
        spine.set_linewidth(2.0)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
